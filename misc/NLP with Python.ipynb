```python
# product some simple stats for gutenberg 
for txt in gutenberg.fileids():
    num_words = len(gutenberg.words(txt))
    avg_word_length = sum([len(w) for w in gutenberg.words(txt)]) / num_words
    avg_sent_length = num_words / len(gutenberg.sents(txt))
    lex_div = num_words / len(set([w.lower() for w in gutenberg.words(txt)]))
    print int(avg_word_length), int(avg_sent_length), int(lex_div), txt
```

```python
# explore stylistics with brown corpus
from nltk.corpus import brown
# print Brown categories
print brown.categories()
print
news_text = brown.words(categories="news")
fdist = nltk.FreqDist([w.lower() for w in news_text])
#sorted(fdist.iteritems(), key=lambda(key, value):(value,key), reverse=True)
modals = ['can','could','may','might','must','shall','should','will','would']
for m in modals:
    print m + ":", fdist[m],

print
sf_text = brown.words(categories="science_fiction")
fdist = nltk.FreqDist([w.lower() for w in sf_text])
wh = ['why','when','what', 'who','where']
for m in wh:
    print m + ":", fdist[m],
print "\n"
# conditional frequency distribution
cfd = nltk.ConditionalFreqDist(
                (genre, word)
                for genre in  brown.categories()
                for word in brown.words(categories=genre))
genres = ['news','religion','hobbies','science_fiction','romance','humor']
cfd.tabulate(conditions=genres, samples=modals)
```

```python
# find how american and citizen feature in inaugural speeches by US Presidents
from nltk.corpus import inaugural
cfd = nltk.ConditionalFreqDist(
            (target, fileid[:4])
            for fileid in inaugural.fileids()
            for w in inaugural.words(fileid)
            for target in ['america','citizen']
            if w.lower().startswith(target)
)
plt.figure(figsize=(15, 8))  # the size you want
cfd.plot()
```

```python
# find most newsworthy days in Brown corpus
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
cfd = nltk.ConditionalFreqDist(
                (genre, word)
                for genre in ['news','romance'] #brown.categories()
                for word in brown.words(categories=genre)             
)
#genres = ['news','romance']
cfd.tabulate(samples=days)
cfd.plot(samples=days)
```

```python
# Lexical resources

from nltk.corpus import words
from nltk.corpus import stopwords
# get words in Alice in Wonderland
alice = gutenberg.words('carroll-alice.txt')
fd = FreqDist(words.words())

def unusual_words(text):
    # get vocab
    text_vocab = set([w.lower() for w in text if w.isalpha()])
    eng_vocab = set([w.lower() for w in words.words() ])
    return text_vocab.difference(eng_vocab)
    
#list(unusual_words(alice))[:10]

def content_fraction(text):
    # measure length of text without stop words; divide this value by 
    # total number of words
    non_stop = [w for w in text if w not in stopwords.words('english')]
    return len(non_stop)/len(text)

content_fraction(alice)
```
