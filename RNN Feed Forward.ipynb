```python
import numpy as np
```

```python
timesteps = 100
input_features = 32
output_features = 64
```

```python
inputs = np.random.random((timesteps, input_features))
state_t = np.zeros((output_features,))
```

```python
W = np.random.random((output_features, input_features))
U = np.random.random((output_features, output_features))
b = np.random.random((output_features,))
```

```python
successive_outputs = []
```

```python
for input_t in inputs:
    output_t = np.tanh(np.dot(W, input_t) + np.dot(U, state_t) + b)
    successive_outputs.append(output_t)
    state_t = output_t

final_output_sequence = np.concatenate(successive_outputs, axis=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import normalize

np.random.seed(420)

hypo_vsm = np.random.randint(100, size=(100, 20)) * np.random.randint(2, size=(100,20))
hyper_vsm = np.random.randint(100, size=(100, 20)) * np.random.randint(2, size=(100,20))

hypo_vsm_l2 = normalize(hypo_vsm)
hyper_vsm_l2 = normalize(hyper_vsm)

features = np.hstack((hypo_vsm_l2 - hyper_vsm_l2, np.square(hyper_vsm_l2 - hypo_vsm_l2)))

y = np.concatenate((np.zeros(50), np.ones(50)))

lr = LogisticRegression(random_state = 0,  max_iter=100)
lr.fit(features, y)
```

```python
features[0]
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(lr, features, y, cv=5, scoring='f1')
```
